for column in num_data:
    skewness = stats.skew(num_data[column])
    skewness_results[column] = skewness

# Display the skewness values
for column, skewness in skewness_results.items():
    if skewness > 0:
        skew_type = "positively skewed"
    elif skewness < 0:
        skew_type = "negatively skewed"
    else:
        skew_type = "approximately symmetric"
    
    print(f"{column}: Skewness = {skewness:.2f} ({skew_type})")

```python
#Square Root Transformation 
```

A square root transformation is a type of data transformation used to stabilize variance and make the data more closely follow a normal distribution, particularly when the data exhibits right skewness (positive skewness). It involves taking the square root of each data point. This transformation is useful when you have positive values and want to reduce the influence of extreme values while maintaining interpretability.

```python
df['YearsSinceLastPromotion'] = np.sqrt(df['YearsSinceLastPromotion'])
```

Replcaing the outlier with mean 

```python



mean_value = final_df['Number Companies Worked'].mean()
std_value = final_df['Number Companies Worked'].std()

# Define a function to replace outliers with the mean
def replace_outliers_with_mean(value):
    if (value < mean_value - 2 * std_value) or (value > mean_value + 2 * std_value):
        return mean_value
    else:
        return value


final_df['Number Companies Worked'] = final_df['Number Companies Worked'].apply(replace_outliers_with_mean)
```

Now lets plot the data and check if the outlier is still present 

```python
plt.figure(figsize=(8, 6)) 
sns.set_style("whitegrid")  
sns.boxplot(final_df["Number Companies Worked"], color='lightblue')  
plt.title("\nOutlier Present in NumCompaniesWorked\n", fontsize=15)
plt.xlabel("\nNumCompaniesWorked", fontsize=13)
plt.tick_params(axis='both', which='major', labelsize=10)
plt.show()
```

Since there are majority of 3 rated employees than our ML algorith will be baised towards it so we will have to do balancing on the data 

We will be using the SMOTE Technique 

```python
# Imports
from imblearn.over_sampling import RandomOverSampler,SMOTENC,SMOTE
# Intilizing the model
oversampling = RandomOverSampler()
X_smote, y_smote = oversampling.fit_resample(X,y)
X_smote = pd.DataFrame(X_smote)
X_smote.columns = ['Age', 'Distance From Home', 'Employee Hourly Rate', 'Number Companies Worked',
       'Employee Last Salary Hike Percent', 'Total Work Experience In Years',
       'Training Times Last Year', 'Experience Years At This Company',
       'ExperienceYearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager', 'EmpEducationLevel',
       'EmpEnvironmentSatisfaction', 'EmpJobInvolvement', 'EmpJobLevel',
       'EmpJobSatisfaction', 'EmpRelationshipSatisfaction',
       'EmpWorkLifeBalance', 'Gender', 'EducationBackground', 'MaritalStatus',
       'EmpDepartment', 'EmpJobRole', 'BusinessTravelFrequency', 'OverTime',
       'Attrition']
X_smote
```

```python
y_smote = pd.DataFrame(y_smote)
y_smote.columns = ["PerformanceRatings"]
y_smote.value_counts(normalize=True)*100
```
